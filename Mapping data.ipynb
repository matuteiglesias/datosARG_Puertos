```python
# Load info, format coordinates column

georref_ports = pd.read_csv('./puertos_for_georref_edit.csv', encoding = 'utf-8').dropna()
georref_ports = georref_ports.merge(Port_total_2014, on = 'Puerto', how = 'left').dropna()
georref_ports['coordinates'] = [[float(i) for i in c.split(',')] for c in georref_ports['location']]
georref_ports['coordinates'] = [l[::-1] for l in georref_ports['coordinates'].values] # reverse lat lon
```

```python
### Make up columns that make a feature: geometry, type and properties

georref_ports['geometry'] = [dict({"type": "Point", "coordinates": c}) for c in georref_ports['coordinates']]
georref_ports['type'] = 'Feature'
georref_ports['properties'] = [dict({"Country": "ARG", "Cantidades": x}) for x in georref_ports['Cantidades']]
```

```python
import ast

geojson_df = pd.DataFrame(columns = ["features"])
geojson_df["features"] = ast.literal_eval(georref_ports[['geometry', 'type', 'properties']].to_json(orient = 'records'))
```
